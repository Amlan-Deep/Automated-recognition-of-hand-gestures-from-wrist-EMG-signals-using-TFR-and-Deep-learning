```python
import os
import pandas as pd
import scipy.io

def extract_to_dataframes(mat_file_path):
    mat = scipy.io.loadmat(mat_file_path)
    emg_data = mat['emg']
    stimulus_data = mat['stimulus']
    emg_df = pd.DataFrame(emg_data)
    stimulus_df = pd.DataFrame(stimulus_data)
    return emg_df, stimulus_df

def process_files_in_folder(folder_path):
    # Iterate through each file in the folder path
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".mat"):
            file_path = os.path.join(folder_path, file_name)
            
            # Extract dataframes from the current mat file
            emg_dataframe, stimulus_dataframe = extract_to_dataframes(file_path)

            # Perform operations on emg_dataframe
            emg_dataframe['stimulus'] = stimulus_dataframe
            emg_dataframe['average'] = emg_dataframe.iloc[:, 0:12].mean(axis=1)
            emg_dataframe = emg_dataframe.drop(emg_dataframe.columns[0:12], axis=1)

            # Split rows into windows
            window_size = 520
            overlap = 470
            num_windows = (len(emg_dataframe) - window_size) // (window_size - overlap) + 1
            window_dfs = []
            stimulus_dfs = []
            start = 0
            
            for i in range(num_windows):
                end = start + window_size
                window_dfs.append(emg_dataframe.iloc[start:end]['average'])
                stimulus_dfs.append(emg_dataframe.iloc[start:end]['stimulus'])
                start += window_size - overlap
            
            # Create a DataFrame with window and stimulus data
            data = {'Window_{}'.format(i + 1): window for i, window in enumerate(window_dfs)}
            data.update({'Stimulus_{}'.format(i + 1): stimulus for i, stimulus in enumerate(stimulus_dfs)})
            output_df = pd.DataFrame(data)

            # Save the DataFrame to a CSV file
            output_folder = r'D:\ML_models_paper\Ninapro\Ninapro_DB4\output_csv'
            os.makedirs(output_folder, exist_ok=True)
            output_csv = os.path.join(output_folder, f"{os.path.splitext(file_name)[0]}_output.csv")
            output_df.to_csv(output_csv, index=False)

# Example usage:
folder_path = r'D:\ML_models_paper\Ninapro\Ninapro_DB4\mat_files'
process_files_in_folder(folder_path)
```

```python
import pandas as pd
import scipy.io
import os

def extract_to_dataframes(mat_file_path):
    mat = scipy.io.loadmat(mat_file_path)
    emg_data = mat['emg']
    stimulus_data = mat['stimulus']
    emg_df = pd.DataFrame(emg_data)
    stimulus_df = pd.DataFrame(stimulus_data)
    return emg_df, stimulus_df

def process_selected_files(file_paths):
    for file_path in file_paths:
        file_name = os.path.splitext(os.path.basename(file_path))[0]  # Extracting the file name
        emg_dataframe, stimulus_dataframe = extract_to_dataframes(file_path)

        # Perform operations on emg_dataframe
        emg_dataframe['stimulus'] = stimulus_dataframe
        emg_dataframe['average'] = emg_dataframe.iloc[:, 0:12].mean(axis=1)
        emg_dataframe = emg_dataframe.drop(emg_dataframe.columns[0:12], axis=1)

        # Split rows into windows
        window_size = 520
        overlap = 470
        num_windows = (len(emg_dataframe) - window_size) // (window_size - overlap) + 1
        
        for i in range(num_windows):
            start = i * (window_size - overlap)
            end = start + window_size

            window_df = pd.DataFrame({
                'Average': emg_dataframe.iloc[start:end]['average'],
                'Stimulus': emg_dataframe.iloc[start:end]['stimulus']
            })

            # Save the window DataFrame to a CSV file
            output_folder = r'D:\ML_models_paper\Ninapro\Ninapro_DB4\output_csv'
            os.makedirs(output_folder, exist_ok=True)
            output_csv = os.path.join(output_folder, f"{file_name}_Window_{i+1}.csv")
            window_df.to_csv(output_csv, index=False)

# Example usage:
file_paths = [
    r'D:\ML_models_paper\Ninapro\Ninapro_DB4\mat_files\S10_E3_A1.mat',
    # Add more file paths as needed
]
process_selected_files(file_paths)
```

```python
import os
import pandas as pd

def transpose_csv_files(input_path):
    # Iterate over all subfolders in the input path
    for root, dirs, files in os.walk(input_path):
        for file in files:
            # Check if the file is a CSV file
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)

                # Read the CSV file into a pandas DataFrame
                df = pd.read_csv(file_path)

                # Transpose the DataFrame
                transposed_df = df.transpose()

                # Save the transposed DataFrame back to the CSV file
                transposed_file_path = os.path.join(root, f'transposed_{file}')
                transposed_df.to_csv(transposed_file_path, index=False)

                print(f"Transposed and saved {file} to {transposed_file_path}")

# Example usage: Replace 'your_input_path' with the path to your main folder
input_path = r'D:\ML_models_paper\Ninapro\Ninapro_DB4\output_concat'
transpose_csv_files(input_path)
```
